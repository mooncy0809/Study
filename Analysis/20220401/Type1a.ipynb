```python
# 2. 분석 요쳥 사항
# 주어진 데이터의 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련 데이터로 추출 후
# train 데이터에서 'total_bedrooms' 변수의 결측값(NaN)을 'total_bedrooms' 변수의 중앙값으로 대체하고,
# 대체전의 'total_bedrooms' 변수 표준편차 값과
# 대체 후의 'total_bedrooms'변수 표준편차 값을 산출한다. 
# 변환 이전과 이후의 표준편차 차이를 파악하여 제출해주세오. 
# Data: ./data/housing.csv  Answer: 1.975147
```

```python
df = pd.read_csv('./data/housing.csv')
df.describe()
```

```python
# 주어진 데이터의 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련 데이터로 추출 후
print(df.shape)
per80 = int(df.shape[0] * 0.8)
print(per80)
train  = df.iloc[0:per80,:]
print(train.shape)
```

```python
# 대체전의 'total_bedrooms' 변수 표준편차 값과
total_bedrooms_before = train['total_bedrooms'].std()
print('total_bedrooms_before: ', total_bedrooms_before)
# train 데이터에서 'total_bedrooms' 변수의 결측값(NaN)을 
# total_bedrooms' 변수의 중앙값으로 대체하고,
median = train['total_bedrooms'].median()
#print(median)
train['total_bedrooms'].fillna(median, inplace=True)
#print(train['total_bedrooms'].isnull().sum()) # --> NaN 발생 횟수 체크

# 대체 후의 'total_bedrooms'변수 표준편차 값을 산출한다. 
total_bedrooms_after = train['total_bedrooms'].std()
print('total_bedrooms_after: ', total_bedrooms_after)
```

```python
#변환 이전과 이후의 표준편차 차이를 파악하여 제출해주세요
print(abs(total_bedrooms_after - total_bedrooms_before))
```

```python
# 3. 분석 요쳥 사항
#    Insurance 데이터 세트에서 charges 항목에서 이상값의 합을 파악하여 제출해주세오.
#    (이상값은 평균에서 1.5 표준편차 이상인 값) 
# Data: ./data/insurance.csv  Answer: 6421430
```

```python
df = pd.read_csv('./data/insurance.csv')
df.head()
```

```python
mean = df['charges'].mean()
std = df['charges'].std()
lower = mean - (std * 1.5)
upper = mean + (std * 1.5)
condition = (df['charges'] < lower) | (df['charges'] > upper)
print(sum(condition))
print(df[condition].shape)
print(df[condition]['charges'].sum())
```

```python
# 4. 분석 요쳥 사항
# 아래와 같이 주어진 airquality  데이터 세트에서 Solar.R에 결측값이 있는 행을 제거하고,
# Ozone 항목의 결측값을 중앙값으로 대체한 후
# 중앙값으로 대체하기 이전과 이후의 Ozone의 표준편차의 차이를 파악하여 제출해주세오.
# 데이터: https://forge.scilab.org/7index.php/p/rdataset/source/tree/master/csv/datasets/airquality.csv
# Data: ./data/airquality.csv  Answer: 3.90557
```

```python
df = pd.read_csv('./data/airquality.csv')
df.head(10)
```

```python
df.describe()
```

```python
# airquality  데이터 세트에서 Solar.R에 결측값이 있는 행을 제거하고,
cdt = df['Solar.R'].isnull()
df = df[-cdt]
print(df.shape)
```

```python
# Ozone 항목의 결측값을 중앙값으로 대체한 후
median = df['Ozone'].median()
std_before = df['Ozone'].std()
print(median)
print(std_before)
df['Ozone'].fillna(median, inplace=True)
std_after = df['Ozone'].std()
print(abs(std_after))
```

```python
# 중앙값으로 대체하기 이전과 이후의 Ozone의 표준편차의 차이를 파악하여 제출해주세오.
print(abs(std_after-std_before))
```

```python
# 5. 분석 요쳥 사항
# 다음과 같이 Hitters 데이터 세트가 주어졌을 때, Salary의 이상값의 합을 파악하여 제출해주세오.
# (단, 이상값은 중위수에서 IQR의 2배를 초과하는 값으로 한다.) 
# Data: ./data/Hitters.csv
```

```python
df = pd.read_csv('./data/Hitters.csv')
df.head()
```

```python
desc = df.describe()
print(desc)
```

```python
# Salary의 이상값의 합을 파악하여 제출
# (단, 이상값은 중위수에서 IQR의 2배를 초과하는 값으로 한다.) 
median = df['Salary'].median()
#print(median)
iqr = desc.loc['75%','Salary']-desc.loc['25%','Salary']
#print(iqr)
lower = median - (iqr*2)
#print(lower)
upper = median + (iqr*2)
#print(upper)
cdt = (df['Salary']<lower) | (df['Salary']>upper)
#print(cdt)
ds = df[cdt]
#print(ds.shape)
print(ds['Salary'].sum())
```

```python
# 값이 없는 경우 형변환 에러를 피할 수 있음
df['TotalCharges'] = df['TotalCharges'].str.replace(" ", "")
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
print(df.dtypes)
```
