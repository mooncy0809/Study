```python
age_sum = [0] * 6
f = open('./sales.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
header = next(data)

for row in data:
    for i in range(1, 7):  # 0은 연도임으로 제외하고 index 1 ~ 6을 사용
        # 20대 기준: age_sum[0] = age_sum[0] + int(row[1])
        # 30대 기준: age_sum[1] = age_sum[1] + int(row[2])
        age_sum[i-1] = age_sum[i-1] + int(row[i]) # index는 0부터 시작

# plt.figure(dpi = 150) # 그래프 크기
plt.figure(figsize=(10, 8)) # 그래프 크기
area = age_sum      # pie plot area
label = header[1:]  # 1~, 'year' 제외
color = ['darkorange', 'gold', 'yellowgreen', 'greenyellow', 'lightcoral', 'lightskyblue']
plt.pie(area, colors=color, labels=label, autopct='%.1f%%',
        startangle=0, shadow=True, textprops={'fontsize': 14})
plt.title('눈꽃 Resort 예약 분포 현황', fontsize=20)
plt.legend(fontsize=12)
plt.axis('equal') # 가로세로 비율이 같은 원 출력
plt.savefig('./sales.png') # 파일 생성
plt.show()            

f.close()
```
