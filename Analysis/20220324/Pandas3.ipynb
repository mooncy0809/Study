```python
code = pd.read_excel('./data/code.xlsx', 
                         header=0, 
                         skipfooter=0, 
                         usecols='A:B')
print(type(code))
code.head(3)
```

```python
# how='inner' df와 code가 서로 대응하는 관계, 국적코드 기준으로 통합
# A31은 국적 코드가 없음으로 출력 안됨, SQL: Inner, Equal
df_code = pd.merge(left=df, 
                   right=code,
                   how='inner',
                   left_on='국적코드',
                   right_on='국적코드')
df_code
```

```python
# how='left' df는 모두 출력되고 code가 대응하는 관계, 국적코드 기준으로 통합
# A31은 국적 코드가 없음, SQL: LEFT OUTER JOIN
df_code = pd.merge(left=df, 
                   right=code,
                   how='left',
                   left_on='국적코드',
                   right_on='국적코드')
df_code
```

```python
# how='right' code는 모두 출력되고 df가 대응하는 관계, 국적코드 기준으로 통합
# df의 A31은 국적 코드가 없음으로 생략, SQL: RIGHT OUTER JOIN
df_code = pd.merge(left=df, 
                   right=code,
                   how='right',
                   left_on='국적코드',
                   right_on='국적코드')
df_code
```

```python
# how='left' df는 모두 출력되고 code가 대응하는 관계, 국적코드 기준으로 통합
# A31은 국적 코드가 없음, SQL: LEFT OUTER JOIN
df_code = pd.merge(left=df, 
                   right=code,
                   how='left',
                   left_on='국적코드',
                   right_on='국적코드')
df_code
```

```python
# pivot table, 그룹화 처리
# values: 집계 데이터
# index: 분류 기준 컬럼, 행분류명
# columns: 컬럼명, 열 분류명
# aggfunc: 데이터에 적용할 함수
df_code_pt = df_code.pivot_table(values='입국객수',
                                 index='국적명',
                                 columns='기준년월',
                                 aggfunc='mean' )
df_code_pt
```

```python
# pivot table, 그룹화 처리
# values: 집계 데이터
# index: 분류 기준 컬럼, 행분류명
# columns: 컬럼명, 열 분류명
# aggfunc: 데이터에 적용할 함수
#index 등 그룹화 컬럼의 값이 NaN이면 집계에서 제외됨
df_code_pt = df_code.pivot_table(values='입국객수',
                                 index='국적명',
                                 columns='기준년월',
                                 aggfunc='max' )
df_code_pt
```
