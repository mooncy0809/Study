## 함수의 사용
- Series, DataFrame에 정밀한 처리 가능

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
ages = Series([10,13,15,23,22,36])
ages
```

```python
def proc(age):
    if age >= 10 and age < 20:
        rg = '10대'
    elif age >= 20 and age < 30:
        rg = '20대'
    elif age >= 30 and age < 40:
        rg = '30대'
    else:
        rg = pd.np.nan
    return rg
```

```python
agesrg = ages.map(proc)
print(type(agesrg))
print(agesrg)
```

```python
print(ages)
```

```python
df = DataFrame([[1,3,5],[4,5,6],[7,8,10]], columns=['A', 'B', 'C'])
print(df)
```

```python
def proc(sr):
    return sr.max()-sr.min()
dfcol = df.apply(proc,axis=0) #A열,B열,C열로 열단위 함수 적용
print(type(dfcol))
print(dfcol)
```

```python
dfrow = df.apply(proc,axis=1) #0행,1행,2행으로 행단위 함수 적용
print(type(dfrow))
print(dfrow)
```

```python
def proc(x):
    return x * x
df['A'] = df['A'].apply(proc)
df
```

```python
# df2 = df.apply(proc) #오류
df2 = df.applymap(proc)
df2
```
